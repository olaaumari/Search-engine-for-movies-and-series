### Traitement des données manquantes dans nos bases de données 

En analyse de données, la première chose à faire est le nettoyage de la base de données, autrement dit le traitement des données manquantes. Ici dans notre cas on a remarqué qu’il existe plusieurs variables qui présentent des données manquantes, la question qui se pose est comment les traiter ces valeurs manquantes ? Doit-on supprimé toutes les données manquantes dans la base ?
Après plusieurs analyses, on a décidé de ne traiter aucune donnée manquante en raison d'une perte d’information dans les bases de données.
Sur notre exemple sur la base de données Amazon en appliquant amazon.isnull()  on voit que la variable country et date_added présente des données manquantes identifié par True. On pouvait aussi faire le contraire c’est-à-dire amazon.notnull()
Amazon.fillna(0) de remplacer les valeurs manquantes par 0 ou de remplacer ces valeurs manquantes au tour par la commande amazon.fillna(method=’bfill’). En général cette méthode n’est pas considérée comme une bonne méthode, car ça cause une transformation de l’information dans la base de données.
On peut également appliquer une autre méthode qui s’appelle amazon.dropna() qui permet d’enlever tous les lignes contenant des données manquantes, ce qui est un peu agressif, on pourra  également sélectionner des colonnes auxquelles on souhaite appliquer la suppression des valeurs manquantes par la méthode amazon.dropna(subset[‘ le nom de la colonne’])

# QUESTION 1
#### Quelle est la plateforme qui contient le plus de diversité géographique de films et séries ? Faire un graphique
Pour répondre à cette question, on commence par l’importation des modules qui nous permet de tracer notre graphique soit seaborn et matplotlib.pyplot. Ici, on s’intéresse à la variable country, qui contient les pays de productions des films et de séries. Pour répondre cette question, on est passé par plusieurs analyses, en regardant la colonne country, on a remarqué qu’un film/série a plusieurs pays de producteur, séparé par des virgules, ce qui implique qu’on risque de compter un pays deux ou plusieurs fois, on a donc décidé d’appliquer la méthode split pour séparer les pays. On a également pris en compte le traitement des données manquantes, pour pouvoir répondre à cette question, car la variable Country est l’une des variables qui contient le plus de données manquantes, vous trouverez en dessous : le programme ainsi que son explication en commentaire. ### Ce qui peut être plus intéressant est de regarder quels sont les pays les plus représentés pour chaque plateforme. Pour se faire on a fait des camemberts pour chaque plateforme.

```python
#on va maintenant afficher des piecharts, non pas pour voir le nombre de pays total représentés sur chaque plateforme, mais les pays les plus représentés sur chaque plateforme. 

#netflix.dropna(subset=['country'], inplace=True)
paysN = netflix_NAN["country"].tolist() 
paysN = ",".join(paysN).split(",")
for i in range(len(paysN)):
     paysN[i] = paysN[i].strip()

#on compte l'occurence de chaque pays
CN = cl.Counter(paysN) 
#on trie par ordre décroissant les occurences. 
sorteNpaysN = sorted(CN.items(), key=operator.itemgetter(1))
#on crée un dictionnaire avec pour clée le nom du pays et pour valeur, le nombre de films ou séries qu'il y a de ce pays sur la plateforme. 
dicoN = dict(sorteNpaysN)

#on crée une liste avec les clés du dictionnaire, et on ne retient que les 5 pays les plus représentés (on prend donc les 5 dernier élements de la liste car ils sont triés par ordre décroissant). 
paysN1 = dicoN.keys()
payslisteN = list(paysN1)
payslisteN = payslisteN[-5:]

#on crée une liste avec les valurs du dictionnaire, et on ne retient que les 5 pays les plus représentés (on prend donc les 5 dernier élements de la liste car ils sont triés par ordre décroissant).
occurrenceN = dicoN.values()
occurrencelisteN = list(occurrenceN)
occurrencelisteN = occurrencelisteN[-5:]

#on créer un piechart pour voir les 5 pays les plus représentés, à l'aide des 2 listes qu'on a crée. 
plt.pie(occurrencelisteN, labels=payslisteN)
plt.axis('on')
plt.legend(title="Catégories")
plt.title ("5 pays les plus représentés sur Netflix")
plt.show()

#Disney

#disney.dropna(subset=['country'], inplace=True)
paysD = disney_NAN["country"].tolist() 
paysD = ",".join(paysD).split(",")
for i in range(len(paysD)):
     paysD[i] = paysD[i].strip()
CD = cl.Counter(paysD) 
sortedpaysD = sorted(CD.items(), key=operator.itemgetter(1))   
dicoD = dict(sortedpaysD)

paysD1 = dicoD.keys()
payslisteD = list(paysD1)
payslisteD = payslisteD[-5:]

occurrenceD = dicoD.values()
occurrencelisteD = list(occurrenceD)
occurrencelisteD = occurrencelisteD[-5:]

plt.pie(occurrencelisteD, labels=payslisteD)
plt.axis('off')
plt.legend(title="Catégories")
plt.title ("5 pays les plus représentés sur Disney Plus")
plt.show()

#Prime_video

#prime_video.dropna(subset=['country'], inplace=True)
paysA = amazon_NAN["country"].tolist() 
paysA = ",".join(paysA).split(",")
for i in range(len(paysA)):
     paysA[i] = paysA[i].strip()
CA = cl.Counter(paysA) 
sorteApaysA = sorted(CA.items(), key=operator.itemgetter(1))   
dicoA = dict(sorteApaysA)

paysA1 = dicoA.keys()
payslisteA = list(paysA1)
payslisteA = payslisteA[-5:]

occurrenceA = dicoA.values()
occurrencelisteA = list(occurrenceA)
occurrencelisteA = occurrencelisteA[-5:]

plt.pie(occurrencelisteA, labels=payslisteA)
plt.axis('off')
plt.legend(title="Catégories")
plt.title ("5 pays les plus représentés sur Prime video")
plt.show()
```
